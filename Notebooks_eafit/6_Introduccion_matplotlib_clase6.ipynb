## Figuras con graficas simples

### Plot()

````python
from numpy import linspace
import matplotlib.pyplot as plt
````

````python
x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.figure(figsize = (5, 5))
plt.plot(x, y, marker = "o") # *, x, o
plt.title("Grafica sin(x)")
plt.xlabel("sin(x)")
plt.ylabel("Amplitud");
````

````python
x = np.linspace(-10, 10, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.figure(figsize = (10, 7))
plt.plot(x, y_sin, marker = "o")
plt.plot(x, y_cos, marker = "x")
plt.plot(x, y_tan, marker = "*", color = "#0BF4CB")
plt.title("Grafica trigonometricas", fontsize = 15)
plt.xlabel("sin(x) cos(x) tan(x)", fontsize = 15)
plt.ylabel("Amplitud", fontsize = 15)
plt.grid(linestyle = "--", linewidth = 0.5, color = "grey")
plt.legend(["sin(x)", "cos(x)", "tan(x)"], loc = "upper left"); # 'upper left','upper right', 'lower left', 'lower right'
````

### Scatter()

Un scatter plot es un grafico de dispersion entre dos variables `una vs la otra`, a continuacion veremos un ejemplo:

````python
x = np.arange(0, 100)
y = x**1.3 + np.random.rand(*x.shape)*30

plt.figure(figsize = (5, 5))
plt.scatter(x, y)

plt.title("Grafica de dispersion", fontsize = 15)
plt.xlabel("Values", fontsize = 15)
plt.ylabel("Frecuencia", fontsize = 15);
````

## Figuras compuestas con subplots()

Aunque las graficas anteriores permiten crear visualizaciones simples, la manera mas comun de encontrarnos la forma de crear dichas graficas es mediante el uso de los `subplots`, el cual nos permite hacer lo siguiente en terminos generales:

- Mas informacion: https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html

### Grafico de dispersion 

````python
x = np.arange(0, 50, 2)
y = x**1.3 + np.random.rand(*x.shape)*30

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 1].plot(x, y)
axs[1, 0].plot(x, y)
axs[1, 1].plot(x, y)
````

````python
fig, axs = plt.subplots(2)
axs[0].plot(x, y)
axs[1].plot(x, y)
````
