## Listar paths

### listdir()

```python
import os
```

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
list_files = os.listdir(path_files)
for i in list_files:
    print(i)
```

### scandir()
Existen otras maneras de listar los directorios, para ellos utilizaremos `os.scandir`, el cual no retorna una lista con todos los archivos y subdirectorios sino que devuelve un `iterador` en lugar de una lista cuando se llama a dicho metodo.

- **Nota:** utilizar este metodo para `Python 3.5 en adelante`

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"

with os.scandir(path_files) as f:
    for files in f:
        print(files)
```

```python
type(files)
```

```python
with os.scandir(path_files) as f:
    for files in f:
        print(files.name)
```

### pathlib

Esta otra forma de listar directorios, la cual proporciona un manejo orientado a objetos para el amnejo de archivos

- **Nota:** utilizar este metodo para `Python 3.4 en adelante`

```python
import pathlib
```

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
files_dir = pathlib.Path(path_files)

for files in files_dir.iterdir():
    print(files.name)
```

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"

for files in pathlib.Path(path_files).iterdir():
    print(files.name)
```

### Listar solo archivos con scandir()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"

with os.scandir(path_files) as files_only:
    for f in files_only:
        if f.is_file():
            print(f.name)
```

### Listar solo archivos con pathlib()

# Crear paths con mkdir()

Python no solo permite listar los paths o directorios que ya se encuentren creados, tambien es posible crear directorios y subdirectorios con el metodo `mkdir`, para ello tambien se hara el uso de `os` y `pathlib`

## os.mkdir()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\test"

os.mkdir(path_files + sub_path)
```

```python
try:
    path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
    sub_path = r"\test"
    os.mkdir(path_files + sub_path)

except FileExistsError as exc: #pythonic
    print("-------------- Exception: --------------")
    print(exc)
```

```python
try:
    path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
    sub_path = r"\test"
    os.mkdir(path_files + sub_path)

except Exception as exc:
    print(exc)
```

```python
### Evitar utilizar un pass dentro del except

try:
    path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
    sub_path = r"\test"
    os.mkdir(path_files + sub_path)

except:
    pass   
```

```python
import sys
import traceback
```

```python
errores = list()
traceback_info = []

try:
    path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
    sub_path = r"\test"
    os.mkdir(path_files + sub_path)

except FileExistsError as exc: #pythonic
    
    errores.append(exc)
    tb = sys.exc_info()[2]
    tbinfo = traceback.format_tb(tb)[0]
    traceback_info.append(tbinfo)
```

```python
errores
```

```python
print(traceback_info[0])
print(errores)
```

## Eliminar archivos

### os.remove()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\test_pathlib\ppt_new.pptx"
path_end = path_files + sub_path

try:
    os.remove(path_end)
except FileNotFoundError as exc:
    print(exc)
```

## pathlib.Path.unlink()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\test_pathlib\Nuevo Archivo WinRAR.rar"
path_end = Path(path_files + sub_path)

try:
    path_end.unlink()
except FileNotFoundError as exc:
    print(exc)
```

## Eliminar directorios

Para eliminar un directorio y no solo un archo especifico se puede utilizar `rm`, hay varias formas de hacerlo, para ello utilizaremos `os`, `pathlib` y `shutil`

### os.rmdir()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\test_pathlib"
trash_dir = path_files + sub_path
os.rmdir(trash_dir)
```

### pathlib.Path.rmdir()

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\test"
path_end = Path(path_files + sub_path)
path_end.rmdir()
```

```python
path_files = r"C:\Users\usuario1\Desktop\Eafit\grupo_DS_NB_G07"
sub_path = r"\Nueva carpeta (2)"
path_end = Path(path_files + sub_path)

try: 
    path_end.rmdir()
except OSError as exc:
    print(exc)    
```
